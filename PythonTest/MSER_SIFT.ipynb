```python
import cv2
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
# filename =r"C:\Users\ZZBZ\Documents\GitHub\FisheyeSFM"
# str1=filename+"\origin_4.jpg"
str1="./origin_4.jpg"

mat=io.ImageCollection(str1)

img=mat[0]
```

```python
mser=cv2.MSER_create()
regions,boxes=mser.detectRegions(img)
kpkp=mser.detect(img)
```

```python
kpkp[0].pt
```

```python

for i in range(len(boxes)):
    x,y,w,h=boxes[i]
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
# cv2.drawKeypoints(img,kpkp,mat[0])
img2 = img.copy()
for marker in kpkp:
    img2 = cv2.drawMarker(img2, tuple(int(i) for i in marker.pt), color=(0, 255, 0))
plt.imshow(img2), plt.show()
```

```python
siftt=cv2.xfeatures2d.SURF_create()
```

```python

kp=siftt.detect(img,None)
##计算kpkp的局部描述子
des=siftt.compute(img,kpkp)
```
